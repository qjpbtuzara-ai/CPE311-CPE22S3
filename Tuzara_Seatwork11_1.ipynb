```python
import matplotlib.pyplot as plt
import numpy as np

b = 1
w1 = 3
x1_single = 0

y_pred_single = b + w1 * x1_single
print(f"Prediction for x1 = {x1_single}: y = {y_pred_single}")

x1_range = np.linspace(-2, 2, 100)
y_range = b + w1 * x1_range

plt.plot(x1_range, y_range, label='$y = 3x + 1$', color='black')
plt.scatter([x1_single], [y_pred_single], color='red', s=100, label='$(0, 1)$', zorder=5)
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.title('Part 1: Simple Linear Regression Prediction')
plt.legend()
plt.grid(True)
plt.savefig('part1_plot.png')
```

I could see that by graphing the line in the form of y = b + mx, with m or the slope being the weight w = 3, x = 0 and b = 1. A point of prediction is at (0,1) and it steps up every 3 units upward. I have also observed that the bias or the y intercept is the baseline on where the weight or slope would start to rise and run.

```python
b = 1
W = np.array([3, -2, 4])


X1 = np.array([3, -1, 2])
X2 = np.array([10, 0, -10])

y1 = b + np.dot(W, X1)
y2 = b + np.dot(W, X2)

print(f"Prediction for Example 1: {y1}")
print(f"Prediction for Example 2: {y2}")

x1_range = np.linspace(-5, 15, 100)
y_range = b + W[0]*x1_range + W[1]*X1[1] + W[2]*X1[2]

plt.figure(figsize=(8, 5))
plt.plot(x1_range, y_range, label='Regression Line (Varying $x_1$, $x_2=-1, x_3=2$)', color='blue')
plt.scatter([X1[0]], [y1], color='red', s=100, label=f'Example 1 point ({X1[0]}, {y1})', zorder=5)


y_range2 = b + W[0]*x1_range + W[1]*X2[1] + W[2]*X2[2]
plt.plot(x1_range, y_range2, label='Regression Line (Varying $x_1$, $x_2=0, x_3=-10$)', color='green')
plt.scatter([X2[0]], [y2], color='darkorange', s=100, label=f'Example 2 point ({X2[0]}, {y2})', zorder=5)


plt.xlabel('$x_1$')
plt.ylabel('Prediction ($y$)')
plt.title('Visualizing Predictions by Varying $x_1$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('proper_plot.png')
```

At part 2 there were two sets of x values which means the x varies. The first one include x1 = 3, x2 = -1, x3 = 2 and the second one being x1 = 10, x2 = 0 and x3 = -10. The points of those being (3,20) and (10, -9) respectively. I have also noticed how positive weights make the prediction go up and negative weights make the prediction go down in which for example a weight of 4 and an x3 = 10 would make the prediction drop at a factor of 40, as a result dragging the y value to -9.
