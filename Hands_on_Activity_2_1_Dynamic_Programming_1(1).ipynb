0/1 Knapsack Problem

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**Code Analysis**

In using the method of recursion, it first states the base case such that if either the capacity w being none or that there are no items left for the recursion to continue, it will return a 0. It will then recursively loop excluding the current item until it finds that the item's weight is less than the capacity then it would then add its value of recursively calling the function with 1 less items only then it will include that item, if not then it would return the maximum value. However, for a bottom up approach using dynamic programming is much simpler as a table could be used to populate iteratively using double for loops for the number of items up from 0 to the max capacity for each subproblem.  In the top-down approach however uses memoization in which the results of the main problem are stored first to avoid recomputation via the calc table, the function then uses recursion to solve the main problem by repeatedly checking the calc table and stores it for later use in order to avoid the aforementioned recalculations.

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
n = int(input("Enter the nth term: "))
memo = {}

def fib(n, memo):
    if n == 0:
        return 0
    if n == 1:
        return 1

    if n not in memo:
        memo[n] = fib(n-2, memo) + fib(n-1, memo)
    return memo[n]
    return memo.values()

print(fib(n, memo))
```

## Supplementary Problem (HOA 2.1 Submission):
* Choose a real-life problem
* Use recursion and dynamic programming to solve the problem

```python
#type your code here for recursion programming solution

prices = [0, 1, 5, 8, 9, 10, 17, 17, 20, 24, 30]

# The length of the raw wood log to cut
log_length = 10

def max_profit_recursive(n):
    # Base Case: If length is 0, no profit
    if n <= 0:
        return 0

    max_val = 0

    # Try every possible first cut (size i)
    for i in range(1, n + 1):
        if i < len(prices):
            current_profit = prices[i] + max_profit_recursive(n - i)

            if current_profit > max_val:
                max_val = current_profit

    return max_val

print(f"Goal: Maximize profit for log length {log_length}\n")
rec_result = max_profit_recursive(log_length)
print(f"Recursive Result: ${rec_result}")
```

```python
#type your code here for dynamic programming solution

memo = {}
log_length = 10

def max_profit_dp(n):
    if n in memo:
        return memo[n]

    if n <= 0:
        return 0

    max_val = 0

    for i in range(1, n + 1):
        if i < len(prices):
            current_profit = prices[i] + max_profit_dp(n - i)

            if current_profit > max_val:
                max_val = current_profit

    memo[n] = max_val
    return max_val

print(f"Goal: Maximize profit for log length {log_length}\n")
dp_result = max_profit_dp(log_length)
print(f"DP Result: $ {dp_result}")
```
