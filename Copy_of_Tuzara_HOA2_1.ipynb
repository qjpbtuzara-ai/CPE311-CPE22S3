```python
n = 4
A = [15, 10, 7, 3]
B = []
C = []

def move_disks(n, source, auxiliary, destination, src, aux, dest):
    if n == 0:
        return
    if n == 1:
        disk = source.pop()
        destination.append(disk)
        print(f"Move disk {disk} from {src} to {dest}")
        print(f"A: {A}, B: {B}, C: {C}")
        print()
        return
    move_disks(n - 1, source, destination, auxiliary, src, dest, aux)

    disk = source.pop()
    destination.append(disk)
    print(f"Move disk {disk} from {src} to {dest}")
    print(f"A: {A}, B: {B}, C: {C}")
    print()

    move_disks(n - 1, auxiliary, source, destination, aux, src, dest)


move_disks(n, A, B, C, 'A', 'B', 'C')
```

**Observations**

I have observed in the algorithm that this uses a recursive algorithm to move the disks which are represented as the elements of an initial list to the destination which is an empty list. The recursion works by calling itself to move n - 1 disks in which case for if n== 1, only 1 disk can be moved from the source to the auxiliary and the destination peg is used temporarily as the auxiliary peg until all the smaller disks are on the auxiliary peg and the largest disk could be moved immedietely to the destination peg in which case only the element 15 is in the source peg. This happens so that a larger disk could not be on top or to the left in this case of the smaller ones. The final recursion happens when the disks are moved from the axuiliary peg to the destination peg in order to move all the smaller disks than 15 to the destination using the source peg as the temporary peg. This happens until all of the disks are inserted to the destination list.
