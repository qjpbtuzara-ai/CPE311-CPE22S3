**Open Data and Private Data**
1. **Open Data**: The Open Knowledge Foundation describes Open Data as "any content, information or data that people are free to use, reuse, and redistribute without any legal, technological, or social restriction."
2. **Private Data**: Data related to an expectation of privacy and regulated by a particular country/government.

**Structured and Unstructured Data**
1. **Structured Data**: Data entered and maintained in fixed fields within a file or record. Easily entered, classified, queried, and analyzed. Relational databases or spreadsheets.
2. **Unstructured Data**: Lacks organization. Raw data such as photo contents, audio, video, web pages, blogs, books, journals, white papers, PowerPoint presentations, articles, email, wikis, word processing documents, and text in general.

```python
import cv2 #Without colab stuff

webcam = cv2.VideoCapture(0)

if not webcam.isOpened():
    print("Error: Could not open webcam.")
    exit()

print("Press 's' to save an image, or 'q' to quit.")

while True:
    try:
        check, frame = webcam.read()
        if not check:
            print("Failed to grab frame")
            break

        #Display the live feed
        cv2.imshow("Capturing", frame)
        
        key = cv2.waitKey(1)
        
        #Save image when 's' is pressed
        if key == ord('s'):
            cv2.imwrite(filename='saved_img.jpg', img=frame)
            print("Processing image...")
            
            #Read back and process
            img_new = cv2.imread('saved_img.jpg', cv2.IMREAD_GRAYSCALE)
            cv2.imshow("Captured Image", img_new)
            cv2.waitKey(1650)
            
            #Further processing
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            img_resized = cv2.resize(gray, (28, 28))
            cv2.imwrite(filename='saved_img-final.jpg', img=img_resized)
            print("Image saved and resized to 28x28!")
            break

        #Quit when 'q' is pressed
        elif key == ord('q'):
            print("Turning off camera.")
            break

    except KeyboardInterrupt:
        break

#Cleanup
webcam.release()
cv2.destroyAllWindows()
print("Program ended.")
```

```python
# Example of gathering voice data using microphone
# Note: Requires system level libraries (e.g. libportaudio2 on Ubuntu)

import sounddevice as sd
from scipy.io.wavfile import write
import wavio as wv

# Sampling frequency
freq = 44100
# Recording duration
duration = 5

# Start recorder with the given values of duration and sample frequency
recording = sd.rec(int(duration * freq), samplerate=freq, channels=2)

# Record audio for the given number of seconds
sd.wait()

# This will convert the NumPy array to an audio file with the given sampling frequency
write("recording0.wav", freq, recording)

# Convert the NumPy array to audio file
wv.write("recording1.wav", recording, freq, sampwidth=2)
print("Audio saved successfully!")
```

```python
import os
import time
import requests
import io
from PIL import Image
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By


print("Setting up browser...")

# 1. Clean Local Setup (No Headless, No Colab stuff)
options = Options()
# 2. Add your custom User-Agent header (Firefox uses 'set_preference' for this)
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:123.0) Gecko/20100101 Firefox/123.0"
options.set_preference("general.useragent.override", user_agent)


driver = webdriver.Firefox(options=options)
```

```python

def scroll_to_end(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5) 

def getImageUrls(name, totalImgs, driver):
    search_url = "https://www.google.com/search?q={q}&tbm=isch"
    driver.get(search_url.format(q=name))
    img_urls = set()
    img_count = 0
    results_start = 0
    
    while(img_count < totalImgs):
        #scroll_to_end(driver)
        # CHANGED: Options.XPATH back to By.XPATH
        thumbnail_results = driver.find_elements(By.XPATH, "//img[contains(@class, 'YQ4gaf')]")
        totalResults = len(thumbnail_results)
        print(f"Found: {totalResults} search results. Extracting links from {results_start}: {totalResults}")
        
        for img in thumbnail_results[results_start: totalResults]:
            try:
                img.click()
                time.sleep(2)
            except Exception:
                continue
                
            # CHANGED: Options.CSS_SELECTOR back to By.CSS_SELECTOR
            actual_images = driver.find_elements(By.CSS_SELECTOR, 'img.iPVvYb, img.sFlh5c, img.n3VNCb')
            for actual_image in actual_images:
                if actual_image.get_attribute('src') and 'https' in actual_image.get_attribute('src'):
                    img_urls.add(actual_image.get_attribute('src'))
                    
            img_count = len(img_urls)
            if img_count >= totalImgs:
                print(f"Found: {img_count} image links")
                break
        else:
            print("Found:", img_count, "looking for more image links ...")
            try:
                # CHANGED: Options.CSS_SELECTOR back to By.CSS_SELECTOR
                load_more_button = driver.find_element(By.CSS_SELECTOR, ".mye4qd")
                driver.execute_script("document.querySelector('.mye4qd').click();")
            except Exception:
                pass
        results_start = len(thumbnail_results)
    return img_urls

def downloadImages(folder_path, file_name, url):
    try:
        image_content = requests.get(url).content
    except Exception as e:
        print(f"ERROR COULD NOT DOWNLOAD {url} {e}")
        return 
        
    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SAVED {url} AT: {file_path}")
    except Exception as e:
        print(f"ERROR COULD NOT SAVE {url} - {e}")

def saveInDestFolder(searchNames, destDir, totalImgs, driver):
    for name in list(searchNames):
        path = os.path.join(destDir, name)
        if not os.path.isdir(path):
            os.makedirs(path)
        print('Current Path:', path)
        
        totalLinks = getImageUrls(name, totalImgs, driver)
        print('totalLinks found:', len(totalLinks))
        
        if totalLinks is None or len(totalLinks) == 0:
            print('Images not found for', name)
        else:
            for i, link in enumerate(totalLinks):
                file_name = f"image_{i:03}.jpg" 
                downloadImages(path, file_name, link)

# --- Execution ---
searchNames = ['cat']
destDir = './Dataset/'
totalImgs = 5

try:
    saveInDestFolder(searchNames, destDir, totalImgs, driver)
finally:
    driver.quit()
    print("Browser closed. Script finished.")
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

print("Starting Firefox...")
options = Options()

# 1. Dress up like a normal human browser
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:123.0) Gecko/20100101 Firefox/123.0"
options.set_preference("general.useragent.override", user_agent)

driver = webdriver.Firefox(options=options)
url = 'https://www.imdb.com/search/title/?release_date=2017-01-01,2017-12-31&sort=num_votes,desc'

try:
    print("Loading IMDb page...")
    driver.get(url)
    
    print("Scrolling down to load more movies...")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3) 
    

    '''try:
        load_more_btn = driver.find_element(By.CSS_SELECTOR, "button.ipc-see-more__button")
        driver.execute_script("arguments[0].click();", load_more_btn)
        print("Clicked 'Load More' button!")
        time.sleep(3)
    except:
        pass # If there is no button, just ignore and keep going'''
        
    # 3. The page is now fully loaded! Grab the raw HTML from Selenium
    print("Grabbing the fully loaded HTML...")
    html = driver.page_source
    
    # 4. Hand the HTML over to BeautifulSoup to do the reading
    html_soup = BeautifulSoup(html, 'html.parser')
    movie_containers = html_soup.find_all('li', class_='ipc-metadata-list-summary-item')
    
    print("-------------------------")
    print("Movies found:", len(movie_containers))
    print("-------------------------")
    
finally:
    # We have our HTML, so we don't need the browser anymore!
    driver.quit()
    print("Browser closed.")
```

```python
names = []
years = []
imdb_ratings = []
metascores = []
votes = []

print("Starting the scraper...")

# 2. Loop through every single movie in our containers list
for container in movie_containers:
    
    # --- TITLE ---
    name = container.find('h3', class_='ipc-title__text').text
    names.append(name)
    
    # --- YEAR ---
    year = container.find('span', class_='dli-title-metadata-item').text
    years.append(year)
    
    # --- RATING & VOTES ---
    rating_tag = container.find('span', class_='ipc-rating-star')
    
    if rating_tag is not None:
        rating_text = rating_tag.text
        imdb = float(rating_text.split()[0])
        vote_count = rating_text.split()[1].strip('()')
    else:
        imdb = None
        vote_count = None
        
    imdb_ratings.append(imdb)
    votes.append(vote_count)
    
    mscore_tag = container.find('span', class_='metacritic-score-box')
    
    if mscore_tag is not None:
        mscore = int(mscore_tag.text.strip())
    else:
        mscore = None
        
    metascores.append(mscore)

print(f"Successfully scraped {len(names)} movies!")
```

```python
from time import time, sleep
from random import randint
from IPython.display import clear_output
from requests import get
from bs4 import BeautifulSoup
import warnings

pages = ['1', '2', '3', '4', '5']
years_url = ['2017', '2018', '2019', '2020']

# Redeclaring the lists to store data in
names = []
years = []
imdb_ratings = []
metascores = []
votes = []

# Preparing the monitoring of the loop
start_time = time()
requests = 0
headers = {"Accept-Language": "en-US, en;q=0.5", "User-Agent": "Mozilla/5.0"}

print("Starting the massive scraper...")

# For every year in the interval 2017-2020
for year_url in years_url:
    
    # For every page in the interval 1-5 
    for page in pages:
        
        # Make a get request
        response = get('https://www.imdb.com/search/title/?release_date=' + year_url + '&sort=num_votes,desc&page=' + page, headers=headers)
        
        # Pause the loop so we don't spam IMDb's servers and get banned
        sleep(randint(8,15))
        
        # Monitor the requests
        requests += 1
        elapsed_time = time() - start_time
        print('Request: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
        clear_output(wait = True)
        
        # Throw a warning for non-200 status codes
        if response.status_code != 200:
            warnings.warn('Request: {}; Status code: {}'.format(requests, response.status_code))
            
        # Break the loop if the number of requests is greater than expected
        if requests > 72:
            warnings.warn('Number of requests was greater than expected.')
            break
            
        # Parse the content
        page_html = BeautifulSoup(response.text, 'html.parser')
        
        # Select all movie containers using the UPDATED IMDb tags!
        mv_containers = page_html.find_all('li', class_='ipc-metadata-list-summary-item')
        
        # Loop through our updated containers
        for container in mv_containers:
            
            # --- TITLE ---
            try:
                name = container.find('h3', class_='ipc-title__text').text
                names.append(name)
            except AttributeError:
                names.append(None)
            
            # --- YEAR ---
            try:
                year = container.find('span', class_='dli-title-metadata-item').text
                years.append(year)
            except AttributeError:
                years.append(None)
            
            # --- RATING & VOTES ---
            rating_tag = container.find('span', class_='ipc-rating-star')
            if rating_tag is not None:
                rating_text = rating_tag.text
                imdb = float(rating_text.split()[0])
                # Check if votes exist in the string before grabbing them
                if len(rating_text.split()) > 1:
                    vote_count = rating_text.split()[1].strip('()')
                else:
                    vote_count = None
            else:
                imdb = None
                vote_count = None
                
            imdb_ratings.append(imdb)
            votes.append(vote_count)
            
            # --- METASCORE ---
            mscore_tag = container.find('span', class_='metacritic-score-box')
            if mscore_tag is not None:
                mscore = int(mscore_tag.text.strip())
            else:
                mscore = None
                
            metascores.append(mscore)

print(f"Finished! Successfully scraped {len(names)} movies across multiple years and pages.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://myanimelist.net/topanime.php?type=airing'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/120.0.0.0 Safari/537.36'}

print("Step 1: Fetching page...")
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the rows - MAL consistently uses 'ranking-list'
rows = soup.find_all('tr', class_='ranking-list')
print(f"Step 2: Found {len(rows)} rows.")

data = []

for row in rows:
    # --- BETTER TITLE SELECTION ---
    # We look for the <h3> tag inside the row, regardless of its class name
    title_tag = row.find('h3') 
    title = title_tag.text.strip() if title_tag else "N/A"
    
    # --- BETTER INFO SELECTION ---
    # We look for the div containing the word 'information'
    info_tag = row.find('div', class_='information')
    info_text = info_tag.text.strip().replace('\n', ' ') if info_tag else "N/A"
    
    # --- SCORE SELECTION ---
    score_tag = row.find('span', class_='text')
    score = score_tag.text.strip() if score_tag else "N/A"

    # Only add to data if we actually found a title
    if title != "N/A":
        data.append({
            'Title': title,
            'Details': info_text,
            'Score': score
        })

# Create DataFrame
df = pd.DataFrame(data)

# --- THE FIX FOR THE EMPTY FILE ---
if not df.empty:
    # 1. Print the count to confirm Python has the data in its brain
    print(f"Step 3: {len(df)} items found. Writing to file...")
    
    # 2. Force the save
    df.to_csv('top_airing_anime.csv', index=False, encoding='utf-8-sig')
    
    # 3. Final verification
    print("Step 4: File 'top_airing_anime.csv' updated successfully.")
    display(df.head(50))
else:
    print("Step 3: DATA LIST IS EMPTY. The tags might be wrong or the page didn't load.")
```
