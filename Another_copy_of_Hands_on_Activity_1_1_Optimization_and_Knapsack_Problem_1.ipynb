2. Create the following methods inside the Food class:
* A method that returns the value of the food
* A method that returns the cost of the food
* A method that calculates the density of the food (Value / Cost)
* A method that returns a string to display the name, value and calories of the food

```python
class Food(object):
    def __init__(self, n, v, w):
        # Make the variables private
        self.name = n
        self.value = v
        self.calories = w
    def getValue(self):
        return self.value
    def getCost(self):
        return self.calories
    def density(self):
        return self.getValue()/self.getCost()
    def __str__(self):
        return self.name + ': <' + str(self.value)+ ', ' + str(self.calories) + '>'
```

3. Create a buildMenu method that builds the name, value and calories of the food


```python
def buildMenu(names, values, calories):
    menu = []
    for i in range(len(values)):
        menu.append(Food(names[i], values[i],calories[i]))
    return menu
```

4. Create a method greedy to return total value and cost of added food based on the desired maximum cost

```python
def greedy(items, maxCost, keyFunction):
    """Assumes items a list, maxCost >= 0,         keyFunction maps elements of items to numbers"""
    itemsCopy = sorted(items, key = keyFunction,
                       reverse = True)
    result = []
    totalValue, totalCost = 0.0, 0.0
    for i in range(len(itemsCopy)):
        if (totalCost+itemsCopy[i].getCost()) <= maxCost:
            result.append(itemsCopy[i])
            totalCost += itemsCopy[i].getCost()
            totalValue += itemsCopy[i].getValue()
    return (result, totalValue)
```

5. Create a testGreedy method to test the greedy method

```python
def testGreedy(items, constraint, keyFunction):
    taken, val = greedy(items, constraint, keyFunction)
    print('Total value of items taken =', val)
    for item in taken:
        print('   ', item)
```

```python
def testGreedys(foods, maxUnits):
    print('Use greedy by value to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, Food.getValue)
    print('\nUse greedy by cost to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, lambda x: 1/Food.getCost(x))
    print('\nUse greedy by density to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, Food.density)
```

9. Create method to use  Bruteforce algorithm instead of greedy algorithm

```python
def maxVal(toConsider, avail):
    """Assumes toConsider a list of items, avail a weight
       Returns a tuple of the total value of a solution to the
         0/1 knapsack problem and the items of that solution"""
    if toConsider == [] or avail == 0:
        result = (0, ())
    elif toConsider[0].getCost() > avail:
        #Explore right branch only
        result = maxVal(toConsider[1:], avail)
    else:
        nextItem = toConsider[0]
        #Explore left branch
        withVal, withToTake = maxVal(toConsider[1:],
                                     avail - nextItem.getCost())
        withVal += nextItem.getValue()
        #Explore right branch
        withoutVal, withoutToTake = maxVal(toConsider[1:], avail)
        #Choose better branch
        if withVal > withoutVal:
            result = (withVal, withToTake + (nextItem,))
        else:
            result = (withoutVal, withoutToTake)
    return result
```

```python
def testMaxVal(foods, maxUnits, printItems = True):
    print('Use search tree to allocate', maxUnits,
          'calories')
    val, taken = maxVal(foods, maxUnits)
    print('Total costs of foods taken =', val)
    if printItems:
        for item in taken:
            print('   ', item)
```

```python
names = ['wine', 'beer', 'pizza', 'burger', 'fries','cola', 'apple', 'donut', 'cake']
values = [89,90,95,100,90,79,50,10]
calories = [123,154,258,354,365,150,95,195]
foods = buildMenu(names, values, calories)
testMaxVal(foods, 2400)
```

#### Supplementary Activity:

```python
def solve_knapsack_raw():
    items = [
        {"name": "Medical Supplies", "w": 10, "v": 60},
        {"name": "Food Rations", "w": 20, "v": 100},
        {"name": "Water Purifiers", "w": 30, "v": 120}
    ]
    capacity = 50

    def get_all_combinations(arr):
        if not arr:
            return [[]]
        first = arr[0]
        rest = arr[1:]
        combs_without_first = get_all_combinations(rest)
        combs_with_first = []
        for lobby in combs_without_first:
            combs_with_first.append(lobby + [first])
        return combs_without_first + combs_with_first

    all_combos = get_all_combinations(items)

    bf_best_value = 0
    bf_best_items = []

    for combo in all_combos:
        current_weight = sum(item['w'] for item in combo)
        current_value = sum(item['v'] for item in combo)

        if current_weight <= capacity and current_value > bf_best_value:
            bf_best_value = current_value
            bf_best_items = [item['name'] for item in combo]


    greedy_items = sorted(items, key=lambda x: x['v']/x['w'], reverse=True)

    g_total_value = 0
    g_total_weight = 0
    g_chosen_items = []

    for item in greedy_items:
        if g_total_weight + item['w'] <= capacity:
            g_chosen_items.append(item['name'])
            g_total_weight += item['w']
            g_total_value += item['v']

    return bf_best_items, bf_best_value, g_chosen_items, g_total_value

bf_items, bf_val, g_items, g_val = solve_knapsack_raw()

print("--- BRUTE FORCE RESULT ---")
print(f"Items: {bf_items}")
print(f"Total Value: {bf_val}")

print("\n--- GREEDY RESULT ---")
print(f"Items: {g_items}")
print(f"Total Value: {g_val}")
```
