```python
import pandas as pd


#Load
fb = pd.read_csv('fb.csv').assign(ticker='FB')
aapl = pd.read_csv('aapl.csv').assign(ticker='AAPL')
amzn = pd.read_csv('amzn.csv').assign(ticker='AMZN')
nflx = pd.read_csv('nflx.csv').assign(ticker='NFLX')
goog = pd.read_csv('goog.csv').assign(ticker='GOOG')

#Combine
faang = pd.concat([fb, aapl, amzn, nflx, goog])

#Export
faang.to_csv('faang.csv', index=False)
faang.head()
```

### Exercise 2: Data Cleaning and Reshaping
* Convert `date` to datetime and `volume` to integers.
* Sort by `date` and `ticker`.
* Find the seven rows with the highest `volume`.
* Melt the dataframe into long format using `date` and `ticker` as ID variables.

```python
#Type conversion
faang['date'] = pd.to_datetime(faang['date'])
faang['volume'] = faang['volume'].astype(int)

#Sorting
faang = faang.sort_values(by=['date', 'ticker'])

#Seven highest volume rows
top_7_volume = faang.nlargest(7, 'volume')

#Melt to long format
faang_long = faang.melt(
    id_vars=['date', 'ticker'],
    value_vars=['open', 'high', 'low', 'close', 'volume']
)

faang_long.head()
```

### Exercise 3: Web Scraping Hospitals
* Scrape a list of hospitals (Name, Address, Contact Info).
* Save to `hospitals.csv`.
* Load into a dataframe and perform preprocessing.

```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/List_of_hospitals_in_the_Philippines"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
all_tables = pd.read_html(response.text)

combined_list = []

for df in all_tables:
    if 'Name of Hospital' in df.columns and 'Class' in df.columns:
        df = df.rename(columns={'Name of Hospital': 'Hospital Name', 'Location': 'Address'})

        filtered_df = df[df['Class'].str.contains('LGU|DOH', case=False, na=False)].copy()

        combined_list.append(filtered_df[['Hospital Name', 'Address', 'Class']])

master_df = pd.concat(combined_list, ignore_index=True)
final_df = master_df.head(100)
final_df.to_csv('hospitals.csv', index=False)

print(f"Total entries captured: {len(final_df)}")
display(final_df)
```
