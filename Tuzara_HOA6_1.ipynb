```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
print(salaries)
```

Exercise 1: Using the data generated above, calculate the following statistics without importing anything from the statistics module in the standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are obtained when using the statistics module (where possible)


* Mean
* Median
* Mode (hint: check out the Counter in the collections module of the standard library at
https://docs.python.org/3/library/collections.html#collections.Counter)
* Sample variance
* Sample standard deviation






```python
#Mean
mean = sum(salaries)/len(salaries)
print(f"mean: ",mean)

#Median
sorted_sal = sorted(salaries)
n = len(sorted_sal)
mid_index = n // 2
if n % 2 == 1:
        median = sorted_sal[mid_index]
else:
        median = (sorted_sal[mid_index - 1] + sorted_sal[mid_index]) / 2

print(f"median: ", median)

#Mode
mode = {}
count = {}
max_count = 0
for item in salaries:
  if item in count:
    count[item] += 1
  else:
    count[item] = 1

max_count = max(count.values())
for key in count:
    if count[key] == max_count:
        mode[key] = max_count

print(count)

print(f"mode: ", mode)

#Sample Variance
s_var = sum((items - mean) ** 2 for items in salaries) / (len(salaries) - 1)
print(f"sample variance: ", s_var)

#Sample Standard Deviation
s_std = s_var**0.5
print(f"sample standard dev: ",s_std)
```

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:
* Range
* Coefficient of variation
* Interquartile range
* Quartile coefficient of dispersion

```python
#Range
range = max(salaries) - min(salaries)
print(f"range: ",range)
#Coefficient of variation
cv = (s_std/mean) * 100
print(f"Coefficient of Variation: ", cv)
#Interquartile Range
n = len(sorted(salaries))
def get_median(data):
    length = len(data)
    mid = length // 2
    if length % 2 == 1:
        return data[mid]
    return (data[mid - 1] + data[mid]) / 2

q1 = get_median(sorted_sal[:n//2])
q3 = get_median(sorted_sal[(n + 1)//2:])
iqr = q3 - q1
print(f"IQR: ", iqr)

#Quartile Coefficient of Dispersion
qcd = (q3 - q1) / (q3 + q1)
print(f"QCD: ", qcd)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('diabetes.csv')

#1. Identify the column names
print("Column Names:", df.columns.tolist())

#2. Identify the data types of the data
print("\nData Types:\n", df.dtypes)

#3. Display the total number of records
print("\nTotal Records:", len(df))

#4. Display the first 20 records
print("\nFirst 20 Records:")
display(df.head(20))

#5. Display the last 20 records
print("\nLast 20 Records:")
display(df.tail(20))

#6. Change the Outcome column to Diagnosis
df.rename(columns={'Outcome': 'Diagnosis'}, inplace=True)

#7. Create a new column Classification
df['Classification'] = df['Diagnosis'].apply(lambda x: 'Diabetes' if x == 1 else 'No Diabetes')

#8. Create a new dataframe "withDiabetes" that gathers data with diabetes
withDiabetes = df[df['Diagnosis'] == 1].copy()

#9. Create a new dataframe "noDiabetes" that gathers data with no diabetes
noDiabetes = df[df['Diagnosis'] == 0].copy()

#10. Create a new dataframe "Pedia" that gathers data with age 0 to 19
Pedia = df[df['Age'] <= 19].copy()

#11. Create a new dataframe "Adult" that gathers data with age greater than 19
Adult = df[df['Age'] > 19].copy()

#12. Use numpy to get the average age and glucose value
avg_age = np.mean(df['Age'])
avg_glucose = np.mean(df['Glucose'])
print(f"\nAverage Age: {avg_age:.2f}")
print(f"Average Glucose: {avg_glucose:.2f}")

#13. Use numpy to get the median age and glucose value
median_age = np.median(df['Age'])
median_glucose = np.median(df['Glucose'])
print(f"Median Age: {median_age}")
print(f"Median Glucose: {median_glucose}")

#14. Use numpy to get the middle values of glucose and age
mid_age = np.median(df['Age'])
mid_glucose = np.median(df['Glucose'])
print(f"Middle Age: {mid_age}")
print(f"Middle Glucose: {mid_glucose}")

#15. Use numpy to get the standard deviation of the skinthickness
skin_std = np.std(df['SkinThickness'])
print(f"SkinThickness Standard Deviation: {skin_std:.2f}")
```
