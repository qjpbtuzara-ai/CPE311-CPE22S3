Isolate the fare_amount, tip_amount, tolls_amount, and total_amount for the longest trip by distance (trip_distance).

```python
longest_trip = df.loc[df['trip_distance'].idxmax()]
print(longest_trip[['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount']])
```

Observation: The longest trip is first defined using .loc[] and the max index of the trip distance which returns rows corresponding to the longest trip. Specific columns are then selected to get the fare, tip, tolls and total amount by the longest trip.
